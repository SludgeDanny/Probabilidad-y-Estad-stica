```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Datos: valores de la variable aleatoria y sus probabilidades
x = np.array([5,6,7,8,9,10])
p = np.array([0.05, 0.10, 0.20, 0.30, 0.25, 0.10])

#Calculo del valor esperado(media)
esperanza = np.sum(x*p)

#Cálculo de la esperanza el cuadrado
esperanza_cuadrado = np.sum(x**2 * p)

#Calculo de la varianza y desviacion estándar
varianza = esperanza_cuadrado - esperanza**2
desviacion_estandar = np.sqrt(varianza)
```

```python
#Mostrar los resultados en consola
print(f"E(X) = {esperanza:.2f}")
print(f"E(X 2) = {esperanza_cuadrado:.2f}")
print(f"Varianza = {varianza:.2f}")
print(f"Desviación Estandar = {desviacion_estandar:.2f}")

#Crear la grafica tipo stem (tallos y hojas)
plt.stem(x,p, basefmt=" ", linefmt='b-', markerfmt='bo')
plt.title("Distribución de Probabilidad Discreta\ncon Media y Desviacióon Estándar")
plt.xlabel("Valores de la variable aleatoria X")
plt.ylabel("P(X)")
plt.grid(True, linestyle ='--', alpha=0.6)

#Dibujar linea vertical en la media
plt.axvline(x=esperanza, color='red', linestyle='--', label=f'Media (E[X]) = {esperanza:.2f}')

#Resaltar el rango de +1 desviavión estándar
plt.axvspan(esperanza - desviacion_estandar, esperanza + desviacion_estandar, color = 'red', alpha= 0.2, label=f"+1 Desv.Estandar (o = {desviacion_estandar:.2f})")
```

```python
#Mostrar leyenda y limites
plt.legend()
plt.ylim(0, max() + 0.05)
plt.xticks(x)

#Mostrar la gráfica
plt.show()
```
