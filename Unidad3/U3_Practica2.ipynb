```python
x = [5, 6, 7, 8, 9, 10] #Variables posibles de la variable aleatoria
p = [0.05, 0.10, 0.20, 0.30, 0.25, 0.10]  #Probaabilidades correspondientes

#Inicializar una variable para acumular la suma
valor_esperado = 0
#Recorrer los valores y sus probabilidades
for i in range(len(x)):
  xi = x[i]
  pi = p[i]
  producto = xi * pi
  valor_esperado += producto

  print(f"x = {xi}, p = {pi}, x*p = {producto:2f}")

print(f"Valor esperado(Esperanza): {valor_esperado:.2f}")
```
