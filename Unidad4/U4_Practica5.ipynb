```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
```

```python
#Datos de la muestra
data = np.array([118, 125, 130, 128,122,135,126,132,129,120,127,124])
n = len(data)
media = np.mean(data)
desv_std = np.std(data, ddof=1) #desviacion estandar muestral (ddof=1)
alpha = 0.05
df = n - 1

#valor critico de t (dos colores)
t_critico = t.ppf(1-alpha/2, df)

#error
error = t_critico * (desv_std / np.sqrt(n))
```

```python
#Intervalo de confianza
lim_inf = media - error
lim_sup = media + error

print(f"Media muestral = {media:.2f} mmHg")
print(f"DEsviacion estandar = {desv_std:.2f} mmmHg")
print(f"Grados de libertad = {df}")
print(f"valor critico t = {t_critico:.3f}")
print(f"Intervalo de confianza = ({lim_inf:.2f}, {lim_sup:.2f} mmHg")

#Graficar
x = np.linspace(media - 3*desv_std, media + 3*desv_std, 200)
y = t.pdf((x - media) / (desv_std / np.sqrt(n)), df)
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='navy', label='Distibución t de la media muestral')
plt.axvline(x=lim_inf, color='red', linestyle='--', label='Limite inferior')
plt.axvline(x=lim_sup, color='red', linestyle='--', label='Limite superior')
plt.axvline(media, color='green', linestyle='-', label='Media muestral')
plt.title("Intervalo de confianza del 95% para la media (t de Student)")
plt.ylabel("Presión arterial (mmHg)")
plt.legend()
plt.grid(True)
plt.show()
```
