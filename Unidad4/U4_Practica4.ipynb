```python
from scipy.stats import binom

#Parámetros de la distribución binomial
probabilidad_exito = 0.83
num_pacientes = 16
X = 16 #Numero de exitos

#Calcular la probabilidad de que nungún paciente sufra la enfermedad
probabilidad_ningun_enfermo = binom.pmf(X, num_pacientes, probabilidad_exito)

print("La probabilidad de que ningún paciente sufra la enefrmedades: ", probabilidad_ningun_enfermo)
```

```python
from scipy.stats import binom

#Parámetros de la distribución binomial
probabilidad_exito = 0.83
num_pacientes = 16
X = 0

#Calcular la probabilidad de que todos los pacientes sufran la enfermedad
probabilidad_todos_enfermo = binom.pmf(X, num_pacientes, probabilidad_exito)

print("La probabilidad de que todos los pacientes sufran la enefrmedades: ", probabilidad_todos_enfermo)
```
