```python
import matplotlib.pyplot as plt
from scipy.stats import norm

plt.figure(figsize=(6, 4))

#Parametros de la distribución normal
media = 70 #Media
desv_std = 4 #Desviación estándar
total_estudiantes = 500

#VARIABLE para calculo de la probabilidad de que un estudiante que pese menos de 75 kg
peso_limite = 75

#Primero necesitamos calcular la probabilidad de que un estudiante pese menos
#de 75 kg utilizando la funcion de distribucion acumulada (CDF) de la distribucion normal
prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)
```

```python
import numpy as np
#Calculo de número de estudiantes que pesan menos de 75 kg
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg

print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Numero de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)

#Graficar la distribucion normal
x = np.linspace(media - 4 *desv_std, media + 4 *desv_std, 1000)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese menos de 75 kg. \n, u=70 ,o=4 ,X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))

# Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500

# VARIABLE para Cálculo de la probabilidad de que un estudiante pese menos de 75 kg peso_limite = 62
# Primero necesitamos calcular la probabilidad de que un estudiante pese menos
#de 75 kg utilizando La función de distribución acumulativa (CDF) de la distribución normal

prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)

# Cálculo del número de estudiantes que pesan menos de 75 kg
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg

print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("Número de estudiantes que pesan menos de 75 kg:", estudiantes_menos_62kg)
```

```python
#Calcular la distribución normal
x= np.linspace (media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

#Graficar la distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title(r'Probabilidad de que un estudiante pese menos de 62 kg, \n u=70, o=4, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))

#parametros de la distribucion normal
media = 70 #Media
desv_std = 4 #Desviacion estandar
peso_limite = 79 #peso limite
total_estudiantes = 500

#Calculamos la probabilidad de que un estudiante pese menos o igual a 79 kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)


#la probabilidad de que un estudiante que pese mas de 79 kg es complementaria
estudiantes_mas_79kg = total_estudiantes * prob_menos_79kg

print("probabilidad de que un estudiante pese mas de 79kg:", prob_menos_79kg)
print("Numero aproximado de estudiantes que pesan de 79 kg:", estudiantes_mas_79kg)
```
