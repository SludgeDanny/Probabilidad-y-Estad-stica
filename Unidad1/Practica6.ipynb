```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
autos2 = pd.read_csv('autos2.csv', encoding="ISO-8859-1")
```

```python
autos2.head()
```

```python
print(f"Valor máximo de precios: {autos2['price'].max()}")
```

```python
print(f"Valor mínimo de precios: {autos2['price'].min()}")
```

```python
print(f"Valor de la moda de precios: \n{autos2['price'].mode()}")
```

```python
print(f"Valor de la media de precios: \n{autos2['price'].mean()}")
```

```python
print(f"Valor de la mediana de precios: \n{autos2['price'].median()}")
```

```python
print(f"Valor de la desviación estandar de precios: \n{autos2['price'].std()}")
```

```python
print("Estadísticos y cuartiles de precios con describe():")
display(autos2['price'].describe())
```

```python
import numpy as np

# Determinar la cantidad de datos (n)
n = len(autos2['price'])

# Aplicar la regla de Sturges para determinar el número óptimo de bins (k)
k = int(1 + np.log2(n))

print(f"Número de datos: {n}")
print(f"Número óptimo de bins : {k}")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(autos2['price'], bins=k, kde=True)
plt.title('Distribución de Precios de los Autos')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='body-style', y='price', data=autos2)
plt.title('Distribución de Precios por Tipo de Vehículo (Body Style)')
plt.xlabel('Tipo de Vehículo (Body Style)')
plt.ylabel('Precio (USD)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=autos2)
plt.title('Distribución de Precios por Tipo de Tracción (Drive Wheels)')
plt.xlabel('Tipo de Tracción (Drive Wheels)')
plt.ylabel('Precio (USD)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='engine-location', y='price', data=autos2)
plt.title('Distribución de Precios por Ubicación del Motor (Engine Location)')
plt.xlabel('Ubicación del Motor')
plt.ylabel('Precio (USD)')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
