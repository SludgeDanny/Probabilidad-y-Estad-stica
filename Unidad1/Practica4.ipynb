```python
import random
import matplotlib.pyplot as ptl
from statistics import *
```

```python
fabrica1 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(50,100),2)
  fabrica1.append(numero_aleatorio)
print(fabrica1)
```

```python
fabrica2 = []
for i in range(100):
  numero_aleatorio = round(random.uniform(40,93),2)
  fabrica2.append(numero_aleatorio)
print(fabrica2)
```

```python
ptl.boxplot(fabrica1)
ptl.ylabel('Producción en mts.')
ptl.title('Producción de la fabrica 1')
ptl.show()
```

```python
fig, (ax1, ax2) = ptl.subplots(nrows=1, ncols=2, figsize=(9, 4))
bplot1 = ax1.boxplot(fabrica1, vert=True, patch_artist=True)
ax1.set_title('Producción de la fabrica 1')
bplot2 = ax2.boxplot(fabrica2, vert=True, patch_artist=True)
ax2.set_title('Producción de la fabrica 2')
ptl.show()
```

```python
fabrica1.append(150)
print(fabrica1)
```

```python
fig,(ax1,ax2) = ptl.subplots(nrows=1,ncols=2,figsize=(9,4))
bplot1 = ax1.boxplot(fabrica1,vert=True,patch_artist=True)
ax1.set_title('Producción de la fabrica 1')
bplot2 = ax2.boxplot(fabrica2,vert=True,patch_artist=True)
ax2.set_title('Producción de la fabrica 2')
ptl.show()
```

```python
fig,(ax1,ax2) = ptl.subplots(nrows=1,ncols=2,figsize=(9,5))
bplot1 = ax1.boxplot(fabrica1,vert=True,patch_artist=True)
ax1.set_title('Producción de la fabrica 1')
ax1.set_ylim(30, 160)
bplot2 = ax2.boxplot(fabrica2,vert=True,patch_artist=True)
ax2.set_title('Producción de la fabrica 2')
ax2.set_ylim(30, 160)
ptl.show()
```
